```python
import matplotlib.pyplot as plt
import numpy as np

a = [1, 2, 3, 4]
b = [5, 6, 7, 8]

c = np.multiply(a, b)

print(a)
```

```python
print(b)
```

```python
print(c)
```

```python
plt.plot(a,a,"r")
plt.ylabel("values of a")
plt.xlabel("values of a")
plt.title("Graph with single Y axes")
plt.show()
```

```python
plt.plot(a,b,"g")
plt.ylabel("values of b")
plt.xlabel("values of a")
plt.title("Graph with single Y axes")
plt.show()
```

```python
plt.plot(a,c,"b")
plt.ylabel("values of c")
plt.xlabel("values of a")
plt.title("Graph with single Y axes")
plt.show()
```

```python
plt.plot(a,a,"r",a,b,"g",a,c,"b")
plt.ylabel("values of a, b and c")
plt.xlabel("values of a")
plt.title("Graph with three(3) Y axes")
plt.show()
```

```python
plt.figure(1)

plt.subplot(131)
plt.plot(a,a,"r")
plt.ylabel("values of a")
plt.xlabel("values of a")
plt.ylim(0,max(c))

plt.subplot(132)
plt.plot(a,b,"g")
plt.ylabel("values of b")
plt.xlabel("values of a")
plt.ylim(0,max(c))

plt.subplot(133)
plt.plot(a,c,"b")
plt.ylabel("values of c")
plt.xlabel("values of a")
plt.ylim(0,max(c))

plt.title("Graph is Equal Size of Axes")
plt.show()
```
